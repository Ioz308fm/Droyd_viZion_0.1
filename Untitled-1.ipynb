```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pytesseract
from PIL import Image
import numpy as np

# Указываем путь к tesseract (если он не в PATH)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Для Windows


def img_plot(img):
    plt.imshow(img)
    plt.axis('off')  # Скрыть оси
    plt.show()

def img_to_text(img_path, output = True):
    # Загрузка изображения
    image = cv2.imread(img_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Бинаризация (Otsu или адаптивная)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

    # Удаление шума
    denoised = cv2.medianBlur(thresh, 3)

    # Морфологическое замыкание (для склейки текста)
    kernel = np.ones((3, 3), np.uint8)
    processed = cv2.morphologyEx(denoised, cv2.MORPH_CLOSE, kernel)

    # Сохранение временного файла для отладки
    cv2.imwrite('processed.jpg', processed)

    # Передача в Tesseract
    text = pytesseract.image_to_string(processed, lang='rus')
    if output:
        img_plot(mpimg.imread(img_path))
    print(text)
```

```python
import easyocr
img_path = "csTFcl8V4NU.jpg"
# Предобработка
img = cv2.imread(img_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)[1]  # Инвертирование для белого текста

# Распознавание
reader = easyocr.Reader(['ru', 'en'])
result = reader.readtext(thresh, detail=0)

img_plot(mpimg.imread(img_path))
print("Распознанный текст:", result)
```
